```python
import random

def GameStart():
    # Score Board initial values
    HumanPoints = 0
    BotPoints = 0
    TiePoints = 0

    while True:
        # Bot generate a random key out three
        bot = random.choice(['Rock','Paper','Scissor'])
        # print(bot)
        # user key
        human = ""
        while human not in ['Rock','Paper','Scissor']:
            human = input("Please insert your choice : Rock/Paper/Scissor: ")
        # print(human)

        # checking winner of the round
        if human == bot:
            print("Game ties! Bot choice was ",bot)
            TiePoints += 1

        elif human == 'Rock' and bot == 'Paper':
            print('Bot wins : Bot choice was ',bot)
            BotPoints += 1

        elif human == 'Rock' and bot == 'Scissor':
            print('Human wins : Bot choice was ',bot)
            HumanPoints += 1


        elif human == 'Paper' and bot == 'Rock':
            print('Human wins : Bot choice was ',bot)
            HumanPoints += 1

        elif human == 'Paper' and bot == 'Scissor':
            print('Bot wins : Bot choice was ',bot)
            BotPoints += 1

        elif human == 'Scissor' and bot == 'Rock':
            print('Bot wins : Bot choice was ',bot)
            BotPoints += 1 

        else:
            print('Human wins : Bot choice was ',bot)
            HumanPoints += 1

        # ScoreBoard preperation

        def ScoreBoard():
            print('Human'+ " | " + "Ties" + " | " + "Bot")
            print("------------------")
            print( "  " + str(HumanPoints) + "   |   " + str(TiePoints) + "  |  " + str(BotPoints))
            print("------------------")

        ScoreBoard()
        
        # Keep on the game until player asks to exit
        ask = input("Want to play again (Y/N): ")
        if ask == 'n' or ask == 'N': 
            break
    print("\nThanks for playing") 
```

```python
GameStart()
```
