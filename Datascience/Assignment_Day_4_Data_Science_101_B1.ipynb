```python
#Questions 2: How to create a series from a numpy array? 
x = np.arange(1,5)
print(x,type(x))
y = pd.Series(x)
print(y,type(y))
```

```python
#Questions 3: How to convert the index of a series into a column of a dataframe? 
a = {i : i*i for i in range(1,9)}
print(a)
b = pd.Series(a)
print(b)
c = pd.DataFrame(b)
c = c.reset_index()
print(c)
```
