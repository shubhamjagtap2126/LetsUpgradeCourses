```python
# Question 1 = Try 5 Different functions of the String in Python.

string = "Hello world!"
print("index = ", string.index("w"))
print("split = ", string.split())
print("upper =", string.upper() )
print("lower =", string.lower())
print("title =", string.title())
```

```python
# Question 2 = Try 5 Different functions of the List object in Python
lst = ["Apple", "Tea", 123, 10.2, 3]

print("append =", lst.append("Horse"))
print(lst)
print("index =", lst.index(123))
print("insert =", lst.insert(1,"Coffee"))
print("pop =", lst.pop())
print("reverse =", lst.reverse())
print(lst)
```

```python
# Question 3 = Experiment with at least 5 default functions of Dictionary
d = {"Apple":123,"Tea":10.2,3:"Power"}

print(d)
print("keys =", d.keys())
print("values =", d.values())
print("", d.get("Tea"))
#print("", d.)
print("", d.popitem())
print("", d.pop("Tea"))
print(d)
```
