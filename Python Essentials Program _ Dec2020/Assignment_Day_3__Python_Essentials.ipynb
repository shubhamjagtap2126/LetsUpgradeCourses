```python
# Question 1 = Use IF ELSE and ELIF to write a program in python for your Report Cards.
marks = 25 # int(input())

if marks >= 100:
    print("Please check Number")
elif marks > 80:
    print("Received A Grade")
elif marks > 60:
    print("Received B Grade")   
elif marks > 40:
    print("Received C Grade") 
else:
    print("Received D Grade")
```

```python
# Question 2 = Use For Loop to Print Prime Numbers in between 1 to 1000
for i in range(1,1000):
    if i > 1 :
        for j in range(2,i):
            if(i % j== 0):
                break
        else:
            print(i, sep = ",")
```

```python
# Question 3 = Write a program for printing the tables from 1,10 using Nested For Loop
for i in range (1,11):
    for j in range(1,11):
        print(i, " + ", j, " = ", i+j )
```

```python
# Question 4 = Program to Print X Prime Numbers using While Loop starting from 0, and take the Input of X from the user
x = int(input())
k = 3
while k in range(0,x+1):
    i = 2
    while i < k:
        if(k % i == 0):
            break
        i += 1
    else:
        print(k, sep = ",")
    k += 1
```
